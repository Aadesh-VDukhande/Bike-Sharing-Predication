```python
X = df.drop(["cnt"] , axis = 1)
Y = df["cnt"]
```

```python
# Let's import train test split

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3 , random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X.columns
```

```python
X_train[['temp','hum', 'windspeed','casual', 'registered']] = scaler.fit_transform(X_train
                                                                    [['temp','hum', 'windspeed','casual', 'registered']])
```

```python
X_test[['temp','hum', 'windspeed','casual', 'registered']] = scaler.transform(X_test
                                                                    [['temp','hum', 'windspeed','casual', 'registered']])
```

```python
model = GradientBoostingRegressor()
```
